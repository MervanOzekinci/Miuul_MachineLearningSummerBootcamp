```python
#hadi bunu bi fonksyion ile yapalim
def check_df(df):
  print("********head**********")
  print(df.head())
  print("********tail**********")
  print(df.tail())
  print("********shape**********")
  print(df.shape)
  print("********info**********")
  print(df.info())
  print("********columns**********")
  print(df.columns)
  print("********index**********")
  print(df.index)
  print("********describe**********")
  print(df.describe().T)
  print("********nullvalues**********")
  print(df.isnull().values.any())
  print("********null_sums**********")
  print(df.isnull().sum())
check_df(df)
```

##Kategorik Degisken Analizi!!!


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
pd.set_option("display.width",500)
df=sns.load_dataset("titanic")
df.head()
```

```python
df["embarked"].value_counts()
```

```python
df["sex"].unique()
```

```python
df["sex"].nunique()#number unique
```

```python
##ilgili degiskenlerin kategorik mi object mi bool mu oldugunu bir sorgulayalim
cat_cols=[col for col in df.columns if str(df[col].dtypes) in ["category","object","bool"]]
cat_cols
```

```python
#numerik ama kategorik
num_but_cat=[col for col in df.columns if df[col].nunique()<10 and df[col].dtypes in ["int","float32"]]
num_but_cat
```

```python
cat_but_car=[col for col in df.columns if df[col].nunique()>20 and str(df[col].dtypes) in ["category","object"]]
cat_but_car
```

```python
cat_cols =cat_cols+num_but_cat
cat_cols
```

```python
df[cat_cols]
```

```python
df[cat_cols].nunique()
```

```python
[col for col in df.columns if col not in cat_cols]
```

```python
#fonksiyonle hale getirelim

def cat_summary(dataframe, col_name):
  print(pd.DataFrame({col_name : dataframe[col_name].value_counts(),
                      "Ratio":100* dataframe[col_name].value_counts()/len(dataframe)}))
  print("############################################")
cat_summary(df,"sex")
```

```python
for col in cat_cols:
  cat_summary(df,col)
```
